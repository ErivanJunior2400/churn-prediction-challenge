```python
import os
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import PowerTransformer, LabelEncoder, OneHotEncoder, MinMaxScaler, StandardScaler


sns.set_style('whitegrid')

print("Bibliotecas importadas com sucesso!")
```

```python
# Definindo o caminho relativo para os dados brutos
# O '../' significa "voltar um diretório"
path_dados = '../data/01_raw/telco_customer_churn.csv'

# Carregando os dados usando a função para ler Excel
df = pd.read_csv(path_dados, sep =';')

# Configura o pandas para mostrar todas as colunas
pd.set_option('display.max_columns', None)

# Visualizando as 5 primeiras linhas para confirmar
df.head()
```

```python
# Mostra o número de linhas completamente duplicadas
print(f"Número de linhas duplicadas: {df.duplicated().sum()}")
# Verifica se há algum customerID duplicado
print(f"Número de IDs de cliente duplicados: {df['customerID'].duplicated().sum()}")
```

```python
# Nenhum valor nulo, porém não significa que deixaremos de preencher possíveis valores nulos, pois em produção situações inesperadas podem ocorrer.
df.info()
```

```python
#Claramente MonthlyCharges e TotalCharges são do tipo float, mas estão com separador decimal inadequado

# Lista das colunas que precisam de correção
colunas_para_corrigir = ['MonthlyCharges', 'TotalCharges']

for coluna in colunas_para_corrigir:
    # Passo 1: Garantir que a coluna é do tipo string para usar o método .str
    # Passo 2: Substituir todas as vírgulas por pontos
    df[coluna] = df[coluna].astype(str).str.replace(',', '.', regex=False)

    # Passo 3: Converter a coluna para um tipo numérico (float)
    # O 'errors='coerce'' é uma segurança: se algo não puder ser convertido, vira Nulo (NaN)
    df[coluna] = pd.to_numeric(df[coluna], errors='coerce')
df.head()
```

```python
df.info()
```

```python
# verificando SeniorCitizen	
df.SeniorCitizen.value_counts()
```

```python
# Selecionando as colunas numéricas (não irei incluir SeniorCitizen, pois é só uma classificação etária binária 0 ou 1)
num_columns = ['tenure','MonthlyCharges', 'TotalCharges']
```

```python
print("Abaixo, o resumo estatístico delas:")

# .describe() em variáveis numéricas mostra contagem, média, desvio padrão, min, max e os quartis.
# Usamos display() para uma formatação de tabela mais bonita no Jupyter.
display(df[num_columns].describe().round(2))
```

```python
#Porém, eu gosto de olhar o coeficiente de variação (std/mean). Pois, permite analisar quem possui maior variabilidade
descritivo = df[num_columns].describe().round(2)

descritivo.loc['coef_variacao'] = descritivo.loc['std'] / descritivo.loc['mean'].abs()

descritivo.loc['coef_variacao_%'] = (descritivo.loc['coef_variacao'] * 100)
display(descritivo.round(2))
```

```python
cat_columns = list(set(df.columns.to_list()) - set(num_columns) - {'customerID'})

print("--- Análise de Frequência das Variáveis Categóricas com Exceção da CUSTOMERID ---")

# Vamos criar um loop que passa por cada nome de coluna na sua lista
for coluna in cat_columns:

    # Imprime um separador para deixar a saída mais organizada
    print(f"\n==================== {coluna.upper()} ====================")

    # Calcula e imprime a contagem de cada categoria
    print("Contagem de cada categoria:")
    print(df[coluna].value_counts())

    # (Bônus) Calcula e imprime o percentual de cada categoria
    print("\nPercentual de cada categoria (%):")
    print(round(df[coluna].value_counts(normalize=True) * 100, 2))
```

```python
#Selecionar apenas as colunas numéricas para a análise de correlação
df_numericas = df[num_columns]

#Calcular a matriz de correlação
correlacao = df_numericas.corr()

# Criar o gráfico de heatmap para visualizar a matriz
plt.figure(figsize=(12, 10))
sns.heatmap(
    correlacao, 
    annot=True,      # Mostra os valores de correlação no gráfico
    cmap='coolwarm', # Escolhe um mapa de cores "divergente" (quente para positivo, frio para negativo)
    fmt='.2f'        # Formata os números para terem duas casas decimais
)
plt.title('Heatmap de Correlação das Variáveis Numéricas', fontsize=16)
plt.show()
# daqui vemos que nossa variável target (churn) é desbalanceada e teremos que levar isso em consideração mais tarde
# customerID me chamou atenção nas letras após a numeração, mas não irei investigar se existe algum padrão para melhorar a predição
# TotalCharges é MonthlyCharges multiplicado por tenure. Então TotalCharges taria apenas multicolinearidade ao modelo e irei remover
```

```python
# Gere gráficos de barras empilhadas para cada coluna categórica, mostrando porcentagens
for column in cat_columns:
    # Calcule a porcentagem de churn para cada categoria
    churn_percentage = pd.crosstab(df[column], df['Churn'], normalize='index') * 100
    
    # Plote o gráfico de barras empilhadas
    ax = churn_percentage.plot(kind='bar', stacked=True, figsize=(10, 6))
    
    # Configurações de título, rótulos e legenda
    ax.set_title(f'Porcentagem de Churn por {column}', fontsize=16)
    ax.set_xlabel(column, fontsize=12)
    ax.set_ylabel('Porcentagem (%)', fontsize=12)
    
    # Adicione uma legenda personalizada
    handles, labels = ax.get_legend_handles_labels()
    ax.legend(handles, ['Não', 'Sim'], title='Churn')
    
    # Adicione os rótulos de porcentagem no centro das barras
    for container in ax.containers:
        ax.bar_label(container, fmt='%.1f%%', label_type='center')
        
    # Ajuste a rotação dos rótulos do eixo x para melhor legibilidade
    plt.xticks(rotation=45, ha='right')
    
    # Ajuste o layout para evitar cortes
    plt.tight_layout()
    
    # Exiba o gráfico no notebook
    plt.show()
    
    # Feche o plot para liberar a memória
    plt.close()
```

```python
# a partir dos gráficos acima podemos ver que o mais pode influenciar no churn é o método de pagamento e o contrato.
# Pois eles possuem uma correlação que fogem da média esperada em certas variáveis
```

```python
# Vemos que totalCharges possui uma assimetria. Usar o método Yeo-Johnson pode resolver
column = 'TotalCharges'
plt.figure(figsize=(10, 6))

pt = PowerTransformer(method='yeo-johnson')

# Cria o gráfico de distribuição (histograma + KDE)
sns.histplot(pt.fit_transform(df[[column]]), kde=True, bins=30, color='skyblue', edgecolor='black')

# Configurações de título e rótulos
plt.title(f'Distribuição de {column}', fontsize=16)
plt.xlabel(column, fontsize=12)
plt.ylabel('Frequência', fontsize=12)

# Ajusta o layout para evitar que os rótulos sejam cortados
plt.tight_layout()

# Mostra o gráfico na saída do notebook (sem salvar em arquivo)
plt.show()

# Fecha o plot para liberar a memória
plt.close()
```

```python
#vemos que após o box cox podemos até usar a padronização dos dados. As outras duas variáveis vamos normalizar (0 a 1)
```

```python
# Usarei label encoding para as categorias com apenas dois rótulos (com exceção de contract que claramente tem hierarquia), pois não irá causar problema de hiearquia 
# Porém, para as categorias com mais de 3 rótulos irei usar one-hot enconding para evitar esse problema
colunas_para_remover = ['customerID']
colunas_label_encoding = ['Dependents', 'Partner', 'PhoneService', 'PaperlessBilling', 'gender', 'Contract']
colunas_one_hot_encoding = ['OnlineSecurity', 'StreamingMovies', 'OnlineBackup', 'StreamingTV', 'MultipleLines', 'DeviceProtection', 'TechSupport', 'InternetService', 'PaymentMethod']
coluna_alvo = 'Churn'
colunas_numericas = ['tenure', 'MonthlyCharges', 'TotalCharges']
```

```python
# Remove a coluna customerID
if all(col in df.columns for col in colunas_para_remover):
    df = df.drop(columns=colunas_para_remover, axis=1)

# Converte colunas numéricas para o tipo numérico e trata os valores ausentes
for col in ['MonthlyCharges', 'TotalCharges']:
    df[col] = pd.to_numeric(df[col], errors='coerce')
    if df[col].isnull().sum() > 0:
        valor_mediana = df[col].median()
        df[col].fillna(valor_mediana, inplace=True)
        print(f"Valores nulos em '{col}' preenchidos com a mediana: {valor_mediana}")

# --- 3. Engenharia de Atributos e Codificação ---

# Inicializa um novo DataFrame para os dados processados
df_processado = pd.DataFrame()

# A) Aplica Label Encoding às colunas especificadas
for col in colunas_label_encoding:
    codificador_rotulos = LabelEncoder()
    df_processado[col] = codificador_rotulos.fit_transform(df[col])
    print(f"Label Encoding aplicado a '{col}'. Categorias: {codificador_rotulos.classes_} -> Mapeado para: {codificador_rotulos.transform(codificador_rotulos.classes_)}")

# B) Aplica One-Hot Encoding
codificador_ohe = OneHotEncoder(handle_unknown='ignore', sparse_output=False)
transformado_ohe = codificador_ohe.fit_transform(df[colunas_one_hot_encoding])
# === MELHORIA: Converte o resultado do One-Hot Encoding para o tipo inteiro ===
transformado_ohe = transformado_ohe.astype(int)
# ==============================================================================
df_ohe = pd.DataFrame(transformado_ohe, columns=codificador_ohe.get_feature_names_out(colunas_one_hot_encoding), index=df.index)
df_processado = pd.concat([df_processado, df_ohe], axis=1)
print("\nOne-Hot Encoding aplicado às colunas especificadas e convertido para o tipo inteiro.")

# C) Normalização de Atributos Numéricos
# C.1) MinMaxScaler para 'tenure' e 'MonthlyCharges'
escalador_minmax = MinMaxScaler(feature_range=(0, 1))
df_processado[['tenure', 'MonthlyCharges_escalado']] = escalador_minmax.fit_transform(df[['tenure', 'MonthlyCharges']])
print("\nMinMaxScaler aplicado a 'tenure' e 'MonthlyCharges'.")

# C.2) Transformação de Potência seguida por Padronização para 'TotalCharges'
transformador_potencia = PowerTransformer(method='yeo-johnson')
total_charges_transformado = transformador_potencia.fit_transform(df[['TotalCharges']])
padronizador = StandardScaler()
total_charges_padronizado = padronizador.fit_transform(total_charges_transformado)
df_processado['TotalCharges_padronizado'] = total_charges_padronizado
print("\nTransformação Yeo-Johnson seguida de StandardScaler aplicada a 'TotalCharges'.")

# D) Adiciona a coluna SeniorCitizen e a variável alvo
df_processado['SeniorCitizen'] = df['SeniorCitizen']
df_processado[coluna_alvo] = df[coluna_alvo].map({'Yes': 1, 'No': 0})
```
